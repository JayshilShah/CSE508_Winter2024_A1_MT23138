```python
# Import the necessary libraries
import os
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pickle
```

```python
# Get the list of all the files
all_files = os.listdir("text_files/")
```

### Preprocessing

```python
StopWords = stopwords.words('english')
```

```python
# Function to remove the stopwords from the text
def remove_stopwords(text):
    tokens = [] 
    for token in text:
        if token not in StopWords:
            tokens.append(token)
    return tokens
```

```python
# Function to remove the punctuation from the text
def remove_punc(txt):
    txt_nopunc = "".join([x for x in txt if x not in string.punctuation])
    return txt_nopunc
```

```python
# Function to remove white space tokens if any
def remove_white_space(tokens):
    words = []
    for i in tokens:
        if i != "" and i != " ":
            words.append(i)
    return words
```

```python
counter = 0
for i in all_files:
    counter += 1
    f = open("text_files/"+i, "r")    # opening the file
    strg = f.read()                   # reading the file
    f.close()                         # closing the file
    
    if counter <= 5:                  # to print pre-processing steps for 5 files
        print(f"File {counter} :")
        print("Original string :")
        print(strg)
        lowerCase = strg.lower() 
        print("Lower Case string :")
        print(lowerCase)
        removePunctuations = remove_punc(lowerCase)
        print("String after removing the punctuations :")
        print(removePunctuations)
        tokens = word_tokenize(removePunctuations)
        print("Tokens :")
        print(tokens)
        removeStopwords = remove_stopwords(tokens)
        print("Tokens after removing the stopwords :")
        print(removeStopwords)
        finalTokens = remove_white_space(removeStopwords)
        print("Final tokens after removing the white space tokens :")
        print(finalTokens)
        print()
    else:
        lowerCase = strg.lower() 
        removePunctuations = remove_punc(lowerCase)
        tokens = word_tokenize(removePunctuations)
        removeStopwords = remove_stopwords(tokens)
        finalTokens = remove_white_space(removeStopwords)
        
    file = open('Preprocessed/'+i,'w')       # creating new pre-processed files
    for item in finalTokens:
        file.write(item+"\n")
    file.close()
```

### Inverted Index

```python
new_files = os.listdir("Preprocessed/")
```

```python
strg=''
inverted_index = {}
for i in new_files:
    f = open("Preprocessed/"+i, "r")    # opening the file
    strg = f.read()                   # reading the file
    f.close()                         
    x=strg.split("\n")                # to remove \n as every tokens are in new line
    x = [token for token in x if token != '']   # to remove the space if any after reloading the pre-processed files
    for term in x:
        if term not in inverted_index.keys():
            inverted_index[term] = [i]
        elif term in inverted_index.keys() and i not in inverted_index[term]:
            inverted_index[term].append(i)
```

```python
def preProcessingQuery(query):
    lowerCase = query.lower() 
    removePunctuations = remove_punc(lowerCase)
    tokens = word_tokenize(removePunctuations)
    removeStopwords = remove_stopwords(tokens)
    finalTokens = remove_white_space(removeStopwords)
    return finalTokens
```

```python
def querySearching(finalTokens, ops):
    sets = []                          # to contain list of documents 
    for i in finalTokens:
        # Check if the token is in the inverted index
        if i in inverted_index:
            # If yes, append the corresponding inverted index set to the list
            sets.append(inverted_index[i])
        else:
            print(f"Token {i} is not in the dictionary.")
            return 0
    aux = []
    total = len(sets)
    counter = 0
    # Iterate over the inverted index sets and perform boolean operations
    for i in sets:
        counter+=1
        if counter<=total:
            if counter == 1:
                aux = list(set(aux).union(set(sets[counter-1])))       # to add list of 1st term to the set
            else:
                if ops[counter-2]=='AND':
                    aux = list(set(aux).intersection(set(sets[counter-1])))
                elif ops[counter-2]=='OR':
                    aux = list(set(aux).union(set(sets[counter-1])))
                elif ops[counter-2]=='AND NOT':
                    aux = list(set(aux).difference(set(sets[counter-1])))
                elif ops[counter-2]=='OR NOT':
                    aux1 = list(set(all_files).difference(set(sets[counter-1])))
                    aux = list(set(aux).union(set(aux1)))
                else:
                    print("Not a valid boolean operator.")
                    break
    return aux
```

```python
inputs = int(input("How many queries do you want to enter?"))
results = []
query_strings = []
for i in range(inputs):
    query = input("Query: ")
    operatoins = input("Operations: ")
    query_tokens = preProcessingQuery(query)
    operatoins = operatoins.upper()               # to handle case sensitivity
    ops_tokens = operatoins.split(",")
    # Check if the number of query tokens matches the expected number of operations
    if len(query_tokens) == len(ops_tokens)+1: 
        searchResult = querySearching(query_tokens, ops_tokens)
        # Check if the search result is valid (not equal to 0)
        if searchResult == 0:
            break
        else:
            results.append(searchResult)
        # Construct a string representation of the query for display
        query_string = ''
        for j in range(len(query_tokens)):
            if j==0:
                query_string = query_tokens[j]+' ' 
            else:
                query_string = query_string+ops_tokens[j-1]+' '+query_tokens[j]+' '
            j+=1
        query_strings.append(query_string)
    else:
        print(f"If tokens are {len(query_tokens)} then operations should be {len(query_tokens)-1}")
        break
```

### Positional Index

```python
positional_index = {}
for i in new_files:
    f = open("Preprocessed/"+i, "r")    # opening the file
    strg = f.read()                   # reading the file
    f.close() 
    x=strg.split("\n")
    x = [token for token in x if token != '']
    # Initialize a counter to keep track of the position of each term in the document
    counter = 0
    # Iterate over each term in the list of tokens
    for term in x:
        counter += 1
        # Check if the term is already a key in the positional index
        if term not in positional_index.keys():           # to create dictionaries inside dictionaries
            d = {}
            d[i] = [counter]
            positional_index[term] = d
        else:
            # If the term is already a key, check if the document is a key in the inner dictionary
            if i not in positional_index[term].keys():
                # If not, add the document as a key with the current position as the value
                positional_index[term][i] = [counter]
            else:
                # If the document is already a key, append the current position to the list of positions
                positional_index[term][i].append(counter)
```

```python
def phraseQuerySearch(aux, p_query_tokens):
    # Initialize an empty list to store the final result of phrase query search
    final_p_result = []
    for i in aux:
        # Initialize an empty list to store the cross results for each adjacent pair of query tokens
        p_cross_result = []
        # Iterate over each adjacent pair of query tokens
        for j in range(len(p_query_tokens)-1):
            # Retrieve positional postings for the current and next query tokens in the current document
            a = positional_index[p_query_tokens[j]][i]
            b = positional_index[p_query_tokens[j+1]][i]
            cross = []
            for elem_a in a:
                for elem_b in b:
                    cross.append((elem_a, elem_b))
            cross = [k for k in cross if k[1]-k[0] == 1]
            p_cross_result.extend(cross) 
        p_temp = ()
        for x in p_cross_result:
            p_temp = p_temp + x
        p_temp_merge = list(set(p_temp))
        p_temp_merge.sort()
        res = [p_temp_merge[i + 1] - p_temp_merge[i] for i in range(len(p_temp_merge)-1)]
        # Check if all differences are 1 (indicating adjacent positions)
        if len(set(res)) == 1 and list(set(res))[0] == 1:
            final_p_result.append(i)
    return final_p_result
```

```python
inputs = int(input("How many queries do you want to enter?"))
p_results = []
for i in range(inputs):
    phrase_query = input("Query: ")
    # Preprocess the query to obtain query tokens
    p_query_tokens = preProcessingQuery(phrase_query)
    p_sets = []
    # Iterate over each query token
    for j in p_query_tokens:
        # Check if the token is in the inverted index
        if j in inverted_index:
            # If yes, append the corresponding inverted index set to the list
            p_sets.append(inverted_index[j])
        else:
            print(f"Token {j} is not in the dictionary.")
            break
    aux = []
    for k in range(len(p_sets)):
        if k == 0:
            aux = list(set(aux).union(set(p_sets[k])))
        else:
            aux = list(set(aux).intersection(set(p_sets[k])))
    # Perform phrase query search using the obtained inverted index sets and query tokens
    p_search_results = phraseQuerySearch(aux, p_query_tokens)
    p_results.append(p_search_results)
```
